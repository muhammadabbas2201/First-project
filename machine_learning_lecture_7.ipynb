```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Pandas/Titanic-Dataset.csv')
```

```python
 plt.figure(figsize=(10,6))
sns.countplot(x='Pclass',hue= 'Survived',data = df)
plt.title("Survival rate by passenger class")
plt.xlabel('Passenger class')
plt.ylabel('count')
plt.legend(labels=["Did not survive",'Survived'])
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x='Sex',hue='Survived',data = df)
plt.title("Survival rate by Gender")
plt.xlabel('Gender')
plt.ylabel('count')
plt.legend(labels=['Did not survive','Survived'])
plt.show()
```

```python
# Age distribution of passengers
plt.figure(figsize=(12,8))
sns.histplot(data=df,x='Age', bins=30 ,kde = True)
plt.title("Age distributoin of passenger")
plt.xlabel('Age')
plt.ylabel('count')
plt.axvline(df['Age'].mean(),color = 'red',linestyle = '--',label = 'Mean Age')
plt.legend()
plt.show()
```

```python
sns.histplot(df['Parch'],kde = False)
plt.show()
```

```python
sns.kdeplot(df['Age'])
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.distplot(df['Age'], hist = True)
plt.show()
```

```python
sns.relplot(x = 'Age',y ='Fare', col = 'Pclass', hue = 'Sex',style = 'Sex',kind = 'scatter',data = df)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.scatterplot(x = 'Age',y=  'Fare',hue = 'Sex', data = df)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x = 'Sex',y='Survived',hue ='Pclass',data = df)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.stripplot(x = 'Sex',y='Age',hue = 'Sex',data=df)
plt.show()
```

```python
#For this example, the following version is cleaner and easier to interpret:

plt.figure(figsize=(12,8))

sns.stripplot(
    x='Sex',
    y='Age',
    data=df,
    jitter=True
)

plt.title("Age Distribution by Sex")
plt.xlabel("Sex")
plt.ylabel("Age")

plt.show()
```

```python
sns.boxplot(x ='Survived', y = 'Age', data = df)
plt.show()
```

```python
sns.violinplot(x='Survived',y='Age',hue = 'Sex',data = df)
plt.show()
```

```python
sns.countplot(x = 'Survived', data = df, palette = 'Blues')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.pointplot(x='Survived',y='Sex',hue = 'Pclass',data = df)
plt.show()
```

```python
A = sns.FacetGrid(df, col ='Survived', row = 'Pclass')
A.map(plt.hist,'Age')
A.add_legend()
plt.show()
```
