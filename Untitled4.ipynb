```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(color_codes=True)
```

```python
df = pd.read_csv('/content/data.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df = df.drop(['Vehicle Size'],axis = 1)
```

```python
df.head(5)
```

```python
df = df.rename(columns={'Engine HP':'HP', 'Engine Cylinders':'Cylinders', 'Transmission Type':'Transmission','Driven_Wheels':'Driven Mode','Number of Doors':'Doors'})
df.head(5)
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows:" , duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.count()
```

```python
print(df.isnull().sum())
```

```python
sns.boxplot(x=df["HP"])
```
