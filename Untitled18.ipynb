```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/USA_Housing.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='tab20c_r')
plt.title('Missing data : Training Test')
plt.show()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
x = df.drop('Price', axis = 1)
y = df['Price']
```

```python
from sklearn import preprocessing
pre_process = preprocessing.StandardScaler().fit(x)
x_transform = pre_process.fit_transform(x)
```

```python
x_transform.shape
x_transform
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_transform,y,test_size=0.3,random_state = 101)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
print(y_pred.shape)
print(y_pred)
```

```python
sns.scatterplot(x=y_test,y=y_pred,color = 'blue',label = 'ideal_line')
plt.plot([min(y_test),max(y_test)] , [min(y_test),max(y_pred)], color='red',label = 'regression_line')
plt.legend()
plt.show()
```

```python
results = np.column_stack((y_pred,y_test))
print("Actual values  | Predicted values")
print("-------------------------------")
for actual, predicted in results:
  print(f"{actual: 14.2f} | {predicted: 12.2f}")
```

```python
residual = actual-y_pred.reshape(-1)
print(residual)
```

```python
sns.distplot(residual, kde = True)
```

```python
from sklearn.metrics import mean_squared_error
print("Linear Regression Model")
print('--'*30)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
print("Mean squared error:" ,mse)
print('Root mean squared error:' ,rmse)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor()
rf_regressor.fit(x_train,y_train)
y_pred_rf = rf_regressor.predict(x_test)
DTr = mean_squared_error(y_pred_rf,y_test)
print(DTr)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor()
rf_regressor.fit(x_train,y_train)
y_pred_rf = rf_regressor.predict(x_test)
RFr = mean_squared_error(y_pred_rf,y_test)
print(RFr)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

rf_regressor = GradientBoostingRegressor()
rf_regressor.fit(x_train,y_train)
y_pred_rf = rf_regressor.predict(x_test)
GBR = mean_squared_error(y_pred_rf,y_test)
print(GBR)
```
