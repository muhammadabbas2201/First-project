```python
import pandas as pd
import scipy as stats
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
file_path = '/content/drive/MyDrive/sample_dataset_1000.csv'
df = pd.read_csv(file_path)
```

```python
df.dtypes
```

```python
df_cleaned = df.drop_duplicates()
```

```python
df_cleaned.head()
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned['Salary'].fillna(df_cleaned['Salary'].mean(),inplace = True)
df_cleaned['Age'].fillna(df_cleaned['Age'].mean(), inplace = True)
# inplace
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_cleaned[['Age','Salary']]=scaler.fit_transform(df_cleaned[['Age','Salary']])
df_cleaned[['Age','Salary']].head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_cleaned[['Age','Salary']]=scaler.fit_transform(df_cleaned[['Age','Salary']])
df_cleaned[['Age','Salary']].head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_cleaned['Gender_encoded']=le.fit_transform(df_cleaned['Gender'])
df_cleaned[['Gender','Gender_encoded']].head()
```

```python
import pandas as pd
df_cleaned = pd.get_dummies(df_cleaned,drop_first=True)
print(df_cleaned.head())
```

```python
from scipy.stats import zscore
import numpy as np
z_scores = np.abs(zscore(df_cleaned[['Age','Salary']]))
outliers = (z_scores > 3).any(axis = 1)
df_no_outliers = df_cleaned[~outliers]
df_no_outliers.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
file_path = '/content/drive/MyDrive/sample_dataset_1000.csv'
df= pd.read_csv(file_path)
```

```python
print("\nDataset Preview:")
print(df.head())
```

```python
print(df.columns)
```

```python
df.columns=['ID', 'Name', 'Age', 'Gender', 'Salary', 'Joining Date', 'Is Active',
       'Comments']
print(df.columns)
```

```python
df.dtypes
```

```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()
numeric_cols.remove('ID')
print(numeric_cols)
```

```python
X = df[['Age']]
Y = df['Salary']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
#Display model coefficient
print("\nmodel cofficient")
print(f'intercept:{model.intercept_}')
print(f"slope:{model.coef_[0]}")
```

```python
# Predict for new data
new_ages = pd.DataFrame({'Age':[25,30,35,40,45]})
new_salaries = model.predict(new_ages)
print("\nPredicted Salaries for new_age")
for Age,Salary in zip(new_ages['Age'],new_salaries):
  print(f"Age:{Age},Predicted Salary:{Salary:.2f}")
```
